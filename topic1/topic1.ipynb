**Pandas** — это библиотека Python, предоставляющая широкие возможности для анализа данных. 

Основными структурами данных в Pandas являются классы **Series** (*одномерный индексированный массив данных некоторого фиксированного типа*) и **DataFrame** (*двухмерная структура данных, представляющая собой таблицу, каждый столбец которой содержит данные одного типа; можно представлять её как словарь объектов типа Series*).

Структура DataFrame отлично подходит для представления реальных данных: строки соответствуют признаковым описаниям отдельных объектов, а столбцы соответствуют признакам.

```python
import pandas as pd
import numpy as np
```

Анализируется набор данных по оттоку клиентов телеком-оператора. С помощью функции `read_csv('file_path')` считываются данные из файла формата .csv. Метод `head()` позволяет посмотреть первые 5 строк данных таблицы.

```python
df = pd.read_csv('telecom_churn.csv')
df.head()
```

Свойство (*property*) `shape` класса DataFrame возвращает кортеж, представляющий размерность данных.

```python
df.shape
```

Свойство `columns` возвращает объект типа **Index**, в котором перечислены имена всех столбцов (названия признаков).

```python
df.columns
```

Метод `info()` выводит общую информацию о объекте класса DataFrame – размерность, названия столбцов, их типы, ненулевые значения и объем занимаемой памяти.  

```python
df.info()
```

### Сортировка

Для извлечения отдельного столбца можно использовать кострукцию вида `DataFrame['Name']`.

**Логическая индексация** DataFrame по одному столбцу выглядит следующим образом: `df[P(df['Name'])]`, где `P(df['Name'])` – это некоторое логическое условие, проверяемое для каждого элемента столбца `Name`. Итогом такой индексации является DataFrame, состоящий только из строк, удовлетворяющих условию `P` по столбцу `Name`.

```python
df[df['Churn'] == 1]
```

Вопрос: **Каковы средние значения числовых признаков среди нелояльных пользователей?**

👇

```python
df[df['Churn'] == 1].mean()
```

Скомбинировав предыдущие два вида индексации, ответим на вопрос: **Сколько в среднем в течение дня разговаривают по телефону нелояльные пользователи?**

👇

```python
df[df['Churn'] == 1]['Total day minutes'].mean()
```

Вопрос: **Какова максимальная длина международных звонков среди лояльных пользователей, не пользующихся услугой международного роуминга?**

Если нам нужна первая строчка DataFrame, используем конструкцию `df[:1]` (аналогично срезам для списков и строк).

```python
df[:1]
```

Если нам нужна последняя строчка DataFrame, используем конструкцию `df[-1:]` (также аналогично срезам для списков и строк).

### Применение функций к ячейкам, столбцам и строкам

**Применение функции к каждому столбцу: `apply()`**

К каждому столбцу применяется функция `max` из библиотеки `NumPy` для определения максимального элемента в столбце. Метод `apply()` можно использовать и для того, чтобы применить функцию к каждой строке. Для этого нужно указать параметр `axis=1` – `df.apply(np.max, axis=1)`.

```python
df.apply(np.max)
```

**Применение функции к каждой ячейке столбца: `map()`**

Метод `map()` можно использовать для **замены значений в колонке**, передав ему в качестве аргумента словарь вида `{old_value: new_value}`.

```python
d = {'No': False, 'Yes': True}
df['International plan'] = df['International plan'].map(d)
df.head()
```

Аналогичную операцию можно провернуть с помощью метода `replace`.

```python
df = df.replace({'Voice mail plan': d})
df.head()
```

### Группировка данных

В общем случае группировка данных в Pandas выглядит следующим образом:

`df.groupby(by=grouping_columns)[columns_to_show].function()`

1. К датафрейму применяется метод `groupby`, который разделяет данные по `grouping_columns` – признаку или набору признаков.

2. Выбираются нужные столбцы (`columns_to_show`).

3. К полученным группам применяется функция или несколько функций.

**Группирование данных в зависимости от значения признака `Churn` и вывод статистик по трём столбцам в каждой группе.**

```python
columns_to_show = ['Total day minutes', 
                   'Total eve minutes', 
                   'Total night minutes']

df.groupby(['Churn'])[columns_to_show].describe(percentiles=[])
```

То же самое можно сделать, передав в функцию `agg()` (одним из аргументов принимает одну или несколько функций, которые нужно применить к заданной абсциссе) список функций.

```python
df.groupby(['Churn'])[columns_to_show].agg([np.mean, np.std, np.min, np.max])
```

### Сводные таблицы

Допустим, необходимо посмотреть, как наблюдения в выборке распределены в контексте двух признаков – `Churn` и `International plan`. Для этого можно построить **таблицу сопряженности**, воспользовавшись методом `crosstab()`.

```python
pd.crosstab(df['Churn'], df['International plan'])
```

```python
pd.crosstab(df['Churn'], df['International plan'], normalize=True)
```

В Pandas, так же, как и в Exel, есть такая опция, как **сводные таблицы** (*pivot tables*). В Pandas за сводные таблицы отвечает метод `pivot_table`, который принимает в качестве параметров:

+ `values` – список переменных, по которым требуется рассчитать нужную статистику,

+ `index` – список переменных, по которым нужно сгрупировать данные,

+ `aggfunc` – то, что необходимо посчитать по группам (сумму, среднее, максимум, минимум и др.)

Выведем среднее число дневных, вечерних и ночных звонков для разных Area code:

```python
df.pivot_table(['Total day calls', 'Total eve calls', 'Total night calls'], ['Area code'], aggfunc='mean')
```

### Преобразование датафреймов

Добавить столбцы в DataFrame можно несколькими способами.

Например, необходимо посчитать общее количество звонков для всех пользователей. Для этого создадим объект `total_calls` типа Series и вставим его в датафрейм.

Метод `insert()` вставляет столбец в DataFrame в указанном месте. В данном примере, вставка осуществляется в конец таблицы (`loc=len(df.columns)`).

```python
total_calls = df['Total day calls'] + df['Total eve calls'] + \
              df['Total night calls'] + df['Total intl calls']

df.insert(loc=len(df.columns), column='Total calls', value=total_calls)
df
```

Добавить столбец можно более простым способом, без создания промежуточных Series – с помощью присвоения:

```python
df['Total charge'] = df['Total day charge'] + df['Total eve charge'] + \
                     df['Total night charge'] + df['Total intl charge']
df
```

Для удаления столбцов или строк используется метод `drop()`. В качестве аргументов передаются нужные индексы (списком) и требуемое значение параметра `axis` (`1` – для удаления столбца (признака), `0` – для удаления строки (значение по умолчанию)).

Удалим созданные только что столбцы:

```python
df = df.drop(['Total charge', 'Total calls'], axis=1)
df
```

Строки удаляются следующим образом:

```python
df.drop([1, 2]).head()
```

### Прогнозирование оттока

Посмотрим, как отток связан с признаком "*International plan*". Используем сводную таблицу `crosstab`. При значении аргумента `margins=True` в таблицу добавляются результаты суммирования элементов столбцов, строк и всей таблицы в целом.

```python
pd.crosstab(df['Churn'], df['International plan'], margins=True)
```

Видно, что когда роуминг подключен, доля оттока намного выше. Возможно, большие и плохо контролируемые траты в роуминге очень не нравятся клиентам и приводят к их недовольству телеком-оператором и, соответственно, к их оттоку. 

Посмотрим на ещё один важный признак – "*Customer service calls*". Построим сводную таблицу.

```python
pd.crosstab(df['Churn'], df['Customer service calls'], margins=True)
```

Видно, что доля оттока сильно возрастает начиная с 4х звонков в сервисный центр (смотреть на соотношение значений в столбцах).

Теперь добавим в DataFrame бинарный признак – результат сравнения `Customer service calls > 3`:

```python
df['Many service calls'] = (df['Customer service calls'] > 3).astype('int')

pd.crosstab(df['Many service calls'], df['Churn'], margins=True)
```

Объединим рассмотренные выше условия и построим сводную таблицу для этого объединения и оттока.

```python
pd.crosstab(df['Many service calls'] & df['International plan'], df['Churn'])
```

Таким образом, прогнозируя отток клиентов в случае, когда число звонков в сервисный центр превышало 3 и подключен роуминг (и прогнозируя лояльность – в противном случае), можно ожидать около 85.8% правильных попаданий (ошибка всего в 464 + 9 случаях). Полученные 85.8% неплохая отправная точка (*baseline*) для дальнейшего построения моделей машинного обучения. 

### Резюме:

+ Доля лояльных клиентов в выборке – 85.5%. Самая наивная модель, ответ которой "клиент всегда лоялен" на подобных данных будет угадывать примерно в 85.5% случаев. То есть доли правильных ответов (*accuracy*) последующих моделей должны быть как минимум не меньше, а лучше – значительно выше этого значения;

+ С помощью простого прогноза, который условно можно выразить следующей формулой: `(International plan = True) & (Customer service calls > 3) => Churn = 1, else Churn = 0`, можно ожидать долю угадываний 85.8%, что ещё чуть выше 85.5%.

+ Эти две отправные точки (*baselines*) были получены без всякого машинного обучения. Если окажется, что прикладываемыми усилиями мы увеличиваем долю правильных ответов всего, скажем, на 0.5%, то возможно, мы что-то делаем не так, и достаточно ограничиться простой моделью из двух условий.

+ Перед обучением сложных моделей рекомендуется немного покрутить данные и проверить простые предположения. Более того, в бизнес-приложениях машинного обучения чаще всего начинают именно с простых решений, а потом экспериментируют с их усложнениями.
